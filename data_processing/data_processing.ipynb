```python
import pandas as pd
import numpy as np
import os

df_list = pd.DataFrame()
file_list = os.listdir(os.path.abspath('data')) #데이터 폴더에서 파일 목록 추출

for f in file_list:
    df = pd.read_excel('data/'+f,usecols=[1,2,3,4]) #파일 각 불러오기
    df_list = pd.concat([df_list, df]) #df_list 데이터프레임에 합치기
df_list.columns = ['장소','본문','태그','테마']
df_list.to_excel('./train.xlsx',index=False)
```

# 중복제거

```python
df_list = pd.read_excel('./train.xlsx')
for d in range(len(df_list)):
    df_list.loc[d]['테마'] = list(set(df_list.loc[d]['테마'].replace("'","").replace('[','').replace(']','').replace(' ','').split(',')))
```

```python
df_list.to_excel('./train.xlsx',index=False)
```

# 다중분류

```python
df_list = pd.read_excel('./train.xlsx')
theme_conv = ['스팸','가볼만한곳','가족여행','우정여행','전통','체험','캠핑','관람','맛집','카페']
for theme in theme_conv:
    df_list[theme] = np.where(df_list['테마'].str.contains(theme),1,0)
df_list.to_excel('./multi_train.xlsx',index=False)
```

```python
df_list = pd.read_excel('./multi_train.xlsx')
output_list = ['태그']
for i in ['스팸','가볼만한곳','가족여행','우정여행','전통','체험','캠핑','관람','맛집','카페']:
    File = open(i+".txt","w",-1,"utf-8")
    File.close()
    File = open(i+".txt","a",-1,"utf-8")
    new_df = df_list[output_list + [i]]
    for d in range(1,len(new_df)):
        nnew_list = list(new_df.loc[d])
        text = str(nnew_list[0]).replace("#",'').replace(",",'').replace("'",'') + ',' + str(nnew_list[1]) + '\n'
        File.write(text)
    File.close()
```

# 행추가 (여러개)

```python
df_list = pd.read_excel('./train.xlsx')
```

```python
output_list = []
length =len(df_list)
for i in range(length):
    theme = df_list['테마'][i].replace("'","").replace('[','').replace(']','').replace(' ','').split(',')
    for t in range(len(theme)):
        new_list = list(df_list.loc[i]['장소':'태그'])+[theme[t]]
        output_list.append(new_list)
df = pd.DataFrame(output_list)
df.columns=['장소','본문','태그','테마']
df.to_excel('trainset2.xlsx')
```
